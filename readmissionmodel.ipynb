```python
# readmission_model.py
import pandas as pd
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, precision_score, recall_score
```

```python
# Load dataset (replace with actual path or source)
df = pd.read_csv("hospital_readmission.csv")

# Feature engineering
df["comorbidity_index"] = df[["diabetes", "hypertension", "asthma"]].sum(axis=1)
```

```python
# Preprocessing
features = ["age", "length_of_stay", "comorbidity_index", "num_prior_admissions"]
X = df[features]
y = df["readmitted_within_30_days"]

```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)

# Model training
model = xgb.XGBClassifier(use_label_encoder=False, eval_metric="logloss")
model.fit(X_train, y_train)

```

```python
# Evaluation
y_pred = model.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
```

```python

print("Confusion Matrix:\n", cm)
print(f"Precision: {precision:.3f}")
print(f"Recall: {recall:.3f}")
```
